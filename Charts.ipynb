```python
import matplotlib.pyplot as plt
import random 
```

```python
savings=[0]

for i in range(0, 11):
    savings.append(random.randrange(savings[i], savings[i] + 2500))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```
